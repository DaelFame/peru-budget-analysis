```python
# 01 - Exploración de Datos (EDA)
Este notebook contiene la carga inicial y exploración del dataset de gasto público 2022–2025.
```

```python
import pandas as pd
import os

# Definir ruta de datos
DATA_PATH = "../data/raw/gastos_2022_2025.csv"

# Leer primeras filas para inspección
df = pd.read_csv(DATA_PATH, nrows=1000)  # solo 1000 filas para no saturar memoria
df.head()
```

```python
# Información general del dataset
df.info()

# Número de filas y columnas
print("Filas:", df.shape[0], "Columnas:", df.shape[1])

# Revisar valores faltantes por columna
df.isna().sum().sort_values(ascending=False).head(20)

# Estadísticas básicas de las columnas numéricas
df.describe().T

```

```python
### KPI 1: Tasa de ejecución
**Definición:* Devengado / PIM  
**Significado:** Mide qué porcentaje del presupuesto modificado realmente se gastó.  
**Pregunta que responde:** ¿Qué tan eficiente fue cada sector o departamento en ejecutar su presupuesto?
```

```python
# Crear columnas de tasa de ejecución por año (2022–2025)
for year in range(2022, 2026):
    pim_col = f"PIM_{year}"
    dev_col = f"DEVENGADO_{year}"
    tasa_col = f"TASA_EJEC_{year}"
    
    if pim_col in df.columns and dev_col in df.columns:
        df[tasa_col] = (df[dev_col] / df[pim_col]).replace([float("inf"), -float("inf")], 0).fillna(0)
```

```python
# Ranking de sectores por tasa de ejecución
for year in range(2022, 2026):
    tasa_col = f"TASA_EJEC_{year}"
    if tasa_col in df.columns:
        resumen = (
            df.groupby("SECTOR_NOMBRE", as_index=False)[tasa_col]
              .mean()
              .sort_values(tasa_col, ascending=False)
        )
        print(f"\nTop sectores por tasa de ejecución {year}")
        print(resumen.head(10))
```

```python
### KPI 2: Brecha presupuestal
**Definición:** PIM - Devengado  
**Significado:** Mide cuánto presupuesto quedó sin gastar.  
**Pregunta que responde:** ¿Cuánto dinero asignado no se utilizó en cada sector o región?
```

```python
# KPI 2: Brecha presupuestal (PIM - Devengado)
for year in range(2022, 2026):
    pim_col = f"PIM_{year}"
    dev_col = f"DEVENGADO_{year}"
    brecha_col = f"BRECHA_{year}"
    
    if pim_col in df.columns and dev_col in df.columns:
        df[brecha_col] = df[pim_col] - df[dev_col]
```

```python
### KPI 3: Nivel de certificación
**Definición:** Certificado / PIM  
**Significado:** Indica qué porcentaje del presupuesto fue formalmente comprometido.  
**Pregunta que responde:** ¿Qué tan seguro estaba el gasto planificado frente a lo realmente ejecutado?
```

```python
# KPI 3: Nivel de certificación (Certificado / PIM)
for year in range(2022, 2026):
    pim_col = f"PIM_{year}"
    cert_col = f"CERTIFICADO_{year}"
    certif_col = f"CERTIFICACION_{year}"
    
    if pim_col in df.columns and cert_col in df.columns:
        df[certif_col] = (df[cert_col] / df[pim_col]).replace([float("inf"), -float("inf")], 0).fillna(0)
```

```python
### KPI 4: Nivel de compromiso anual
**Definición:** Comprometido anual / PIM  
**Significado:** Indica cuánto del presupuesto se planificó gastar dentro del año.  
**Pregunta que responde:** ¿Qué tan alineada estuvo la planificación con la ejecución real?
```

```python
# KPI 4: Nivel de compromiso anual (Comprometido anual / PIM)
for year in range(2022, 2026):
    pim_col = f"PIM_{year}"
    comp_col = f"COMPROMETIDO_ANUAL_{year}"
    comp_anual_col = f"COMPROMISO_ANUAL_{year}"
    
    if pim_col in df.columns and comp_col in df.columns:
        df[comp_anual_col] = (df[comp_col] / df[pim_col]).replace([float("inf"), -float("inf")], 0).fillna(0)
```

```python
### KPI 5: Variación interanual
**Definición:** % cambio de PIM y Devengado entre años consecutivos.  
**Significado:** Mide si el presupuesto y la ejecución crecieron o decrecieron año a año.  
**Pregunta que responde:** ¿La ejecución mejora o empeora con el tiempo?
```

```python
# KPI 5: Variación interanual (% cambio entre años)
for year in range(2023, 2026):  # desde 2023 porque comparamos con el año anterior
    pim_prev = f"PIM_{year-1}"
    pim_curr = f"PIM_{year}"
    dev_prev = f"DEVENGADO_{year-1}"
    dev_curr = f"DEVENGADO_{year}"
    
    var_pim_col = f"VAR_PIM_{year}"
    var_dev_col = f"VAR_DEV_{year}"
    
    if pim_prev in df.columns and pim_curr in df.columns:
        df[var_pim_col] = ((df[pim_curr] - df[pim_prev]) / df[pim_prev]).replace([float("inf"), -float("inf")], 0).fillna(0)
    if dev_prev in df.columns and dev_curr in df.columns:
        df[var_dev_col] = ((df[dev_curr] - df[dev_prev]) / df[dev_prev]).replace([float("inf"), -float("inf")], 0).fillna(0)
```
